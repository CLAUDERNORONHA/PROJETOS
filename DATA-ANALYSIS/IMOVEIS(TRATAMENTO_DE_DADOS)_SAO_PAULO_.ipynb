```python
from sklearn.model_selection import train_test_split
```

```python
y = df['Price'] # o meu Y vai ser todo o DF Price
X = df.loc[:, df.columns != 'Price']# e o X vai ser todo o df  for diferente de y Price 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3 )
```

VAMOS PEGAR TRÊS MODELOS PARA TREINAR 

LinearRegression

```python
#LinearRegression


from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

lin_reg.fit(X_train, y_train)#Aqui ele vai utilizar a função custo para fazer a otimização do modelo


#A ideia é a seguinte, treinar os modelos(X_train e y_train) para que fiquem o menor possivel. Que reduz o custo o menor poessivel 

```

```python
#Vamos pegar alguns dados para treinar e ve funcionar 

alguns_dados = X_train.iloc[:5]#esse numero é a quantidade de apartamentos que iremos treinar 
algumas_label = y_train.iloc[:5]#Quantidade de apartamentos

print('predições:', lin_reg.predict(alguns_dados))
print('Labels:', algumas_label.values)


# ESSE FORAM OS VALORES DE TREINO 
#  predições: [1801.05639648 1320.2154541  1254.14489746 2968.69311523 1150.68334961]

#ESSES FORAM OS VALORES VERDADEIROS
# Labels: [1800 1200 2200 3600 1000]
```

```python

from sklearn.metrics import mean_squared_error

preds = lin_reg.predict(X_train)
lin_mse = mean_squared_error(y_train, preds )

#Tirar a raiz quadrada

lin_mse = np.sqrt(lin_mse)
lin_mse
```

A nossa previsão está errando na faixa de R$ 1917 reais  por predição que ele faz.

Ele esta puxando para partamentos mais caros na faixa de 50 mil, por isso o valor ser tão alto nas previsões.

DECISIO TREE REGRESSION 
   - ARVORE DE DECISAO

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(X_train, y_train)
```

```python
preds = tree_reg.predict(X_train)
tree_mse = mean_squared_error(y_train, preds)

tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

Observa que existe uma diferença muito grande entre a arvode de decisão tivemos um valor de R$ 44.

OU seja ele aprendeu muito pouco com a base de treino dele 


Esse modelo não conseguiu compreender o aspecto geral dos dados 

CROSS-VALIDATION

```python
from sklearn.model_selection import cross_val_score

#Pegr a média dos erros quadraddos
scores = cross_val_score(tree_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=10)

tree_rmse_scores = np.sqrt(-scores)#o valor negativo para termos um valor muito parecido que a gente viu anteriormente.

```

```python
#Criar uma função para printar as medias dos Scores

def display_scores(scores):
  print("Scores:", scores)
  print("Mean", scores.mean())
  print("Standard deviation", scores.std())

display_scores(tree_rmse_scores)  
```

Temos um modelos de overfit quando treinamos com omodelo de arvore de decisao
comparando  o modelo de Arvore de decisão co mo modelo crossvalidation, temos uma diferença muito grande
com os valores que ele não viu o cross validations esta errando em 2.274 reais.

O modelo da arvore de decisão treinou demais e ficou burro com a previsão, ou seja, treinou demais os números.

Avaliar e otimizar o MODELO

```python
from sklearn.model_selection import GridSearchCV
```

```python
#Criar uma lista com os paramentos que quero atualizar 

#Vou testar os estimadores 3,10 e 30

param_grid = [
    {'n_estimators':[3,10,30], 'max_features':[2,4,6,8]},
    {'bootstrap':[False], 'n_estimators':[3,10], 'max_features':[2,3,4]}

]







```

```python
#INSTANCIAR UMA FLORESTA ALEATORIA 

forest_reg = RandomForestRegressor()


#passa o modelo que vamos testar, passamos o paramentros 
grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                           scoring = 'neg_mean_squared_error',
                           )
grid_search.fit(X_train, y_train)


#vamos reter o melhor modelo
```

```python
#Vamos pegar o melhor modelo que foi feito na floresta. que foi o melhor resultado

grid_search.best_params_
```

```python
grid_search.best_estimator_
```

Agora vamos saber se o nosso modelo de fato é bom ou não 

```python
, mean_squared_log_error
final_model = grid_search.best_estimator_
final_model_predictions = final_model.predict(X_test)

final_mse = mean_squared_error(y_test, final_model_predictions)
print(np.sqrt(final_mse))
```
