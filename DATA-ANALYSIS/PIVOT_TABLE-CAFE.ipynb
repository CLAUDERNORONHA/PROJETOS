## PIVOT-TABLE()
Estudo de caso - Café

As tabelas dinâmicas são um dos recursos mais poderosos do Excel. Uma tabela dinâmica nos permite extrair insights dos dados. O Pandas fornece uma função semelhante chamada pivot_table(). Pandas pivot_table()é uma função simples, mas pode produzir análises muito poderosas muito rapidamente.
Neste artigo, vamos explorar como usar o Pandas pivot_table()com a ajuda de exemplos. Os exemplos abrangem as seguintes 

* tarefas:

1 Tabela dinâmica mais simples

2-Especificando valuese realizando agregação

3-Vendo quebrar usando columns

4-Substituindo valores ausentes

5-Exibindo valuesvisualização múltipla  

6-AjustandoMostrando total


```python
import pandas as pd
```

```python
cafe = pd.read_csv('coffee_data.csv')
```

```python
cafe.head()
```

```python
#Colocar a região com Index
cafe.pivot_table(index='region')
```

Para exibir varios indices, podemos passar uma lista para index

```python
cafe.pivot_table(index=['region','product_category'])
```

Os valores indexsão as chaves a serem agrupadas na tabela dinâmica. Você pode alterar a ordem dos valores para obter uma representação visual diferente, por exemplo, queremos dar uma olhada nos valores médios agrupando a região com product_category .

```python
cafe.pivot_table(index = ['product_category', 'region'])
```

2. Especificando valores e realizando agregação

Por padrão, pivot_tableexecuta a meanfunção de agregação em todas as colunas numéricas e retorna o resultado. Para especificar explicitamente as colunas de nosso interesse, use o valuesargumento.

```python
cafe.pivot_table(index=['region'],values=['sales'])
```

Para realizar uma agregação diferente de mean, podemos passar uma função de string válida para aggfunc, por exemplo, fazer uma soma:

```python
cafe.pivot_table(index=['region'], values=['sales'], aggfunc='sum' )
```

aggfunc pode ser uma lista de funções, e abaixo está um exemplo para exibir a soma e a contagem

```python
cafe.pivot_table(
index = ['region'],
values = ['sales'],
aggfunc = ['sum','count'])
```

3. Vendo a divisão usando colunas

Se gostaria de ver as vendas discriminadas por PRODUCT_CATEGORY , o columns  argumento nos permite fazer isso

```python
cafe.pivot_table(
index = ['region'],
values = ['sales'],
aggfunc = 'sum',
columns = 'product_category')
```

4. Substituição de valores ausentes

Você provavelmente notou um NaNvalor da saída anterior. Conseguimos isso porque não há vendas de chá no sul. Se quisermos substituí-lo, podemos usar o fill_valueargumento, por exemplo, para definir NaNcomo 0.

```python
cafe.pivot_table(
index = ['region'],
values = ['sales'],
aggfunc = 'sum',
columns = ['product_category'],
fill_value = 0)
```

5. Exibindo vários valores e ajustando a visualização

Se também quisermos ver a soma do custo , podemos adicionar a coluna de custo à valueslista.

```python
cafe.pivot_table(
index = ['region'],
values = ['sales','cost'],
aggfunc = 'sum',
columns = ['product_category'],
fill_value = 0)
```

Isso funciona, mas a representação visual não parece ser útil quando gostaríamos de comparar o custo e as vendas lado a lado. Para obter uma melhor representação visual, podemos mover product_category de columnse adicionar a index.

```python
cafe.pivot_table(
index = ['region','product_category'],
values = ['sales','cost'],
aggfunc ='sum',
fill_value = 0)
```

Agora, a tabela dinâmica é muito conveniente para ver a diferença entre vendas e custos lado a lado.


6. Mostrando total

Outra exibição comum na tabela dinâmica é mostrar o total. No Panda pivot_table(), podemos simplesmente passar margins=True:

```python
cafe.pivot_table(
index = ['region','product_category'],
values = ['sales','cost'],
aggfunc = 'sum',
fill_value = 0,
margins = True)
```
